Trabajando con múltiples figuras y ejes
MATLAB y pyplot tienen el concepto de la figura actual y los ejes actuales. Todas las funciones de trazado se aplican a los ejes actuales. La función `gca` devuelve los ejes actuales (una instancia de matplotlib.axes.Axes), y `gcf` devuelve la figura actual (una instancia de matplotlib.figure.Figure). Normalmente, no tienes que preocuparte por esto, porque todo se maneja automáticamente detrás de escena. A continuación, hay un script para crear dos subgráficos.

La llamada a `figure` aquí es opcional porque se creará una figura si no existe ninguna, al igual que se creará un eje (equivalente a una llamada explícita a `subplot()`) si no existe ninguno. La llamada a `subplot` especifica `numrows`, `numcols`, `plot_number`, donde `plot_number` varía de 1 a `numrows*numcols`. Las comas en la llamada a `subplot` son opcionales si `numrows*numcols<10`. Entonces, `subplot(211)` es idéntico a `subplot(2, 1, 1)`.

Puedes crear un número arbitrario de subgráficos y ejes. Si te resulta molesto que se mantengan los estados (específicamente la imagen, la figura y los ejes actuales) detrás de escena, no te preocupes: esto es solo un envoltorio delgado y con estado alrededor de una API orientada a objetos, que puedes usar en su lugar (consulta el tutorial de Artistas).

Si estás haciendo muchas figuras, debes ser consciente de una cosa más: la memoria requerida para una figura no se libera completamente hasta que la figura se cierra explícitamente con `close`. Borrar todas las referencias a la figura y/o utilizar el administrador de ventanas para cerrar la ventana en la que aparece la figura en la pantalla no es suficiente, porque pyplot mantiene referencias internas hasta que se llama a `close`.

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot() by default

plt.figure(1)                # first figure current;
                            # subplot(212) still current
plt.subplot(211)             # make subplot(211) in the first figure
                            # current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

# Plot types

Los gráficos estadísticos se pueden organizar de varias formas dependiendo de la naturaleza de los datos y la información que se desea resaltar. A continuación, se presentan algunas formas comunes de organizar los tipos de gráficos estadísticos:

1. Gráficos de barras: se utilizan para comparar una variable cuantitativa entre diferentes categorías. Los datos se representan mediante barras verticales o horizontales.
2. Gráficos de líneas: se utilizan para mostrar la evolución de una variable cuantitativa a lo largo del tiempo. Los datos se representan mediante puntos conectados por líneas.
3. Gráficos circulares o de torta: se utilizan para mostrar la distribución porcentual de una variable cuantitativa en diferentes categorías. Los datos se representan mediante porciones de un círculo.
4. Histogramas: se utilizan para mostrar la distribución de frecuencias de una variable cuantitativa continua. Los datos se agrupan en intervalos y se representan mediante barras adyacentes.
5. Gráficos de dispersión: se utilizan para mostrar la relación entre dos variables cuantitativas. Los datos se representan mediante puntos en un plano cartesiano.
6. Gráficos de caja y bigotes: se utilizan para mostrar la distribución de una variable cuantitativa, resaltando su tendencia central y dispersión. Los datos se representan mediante una caja que contiene el 50% central de los datos y líneas que se extienden desde la caja para mostrar la variabilidad.
7. Gráficos de área: se utilizan para mostrar la evolución de una variable cuantitativa a lo largo del tiempo, resaltando la contribución de cada categoría a la variable total. Los datos se representan mediante áreas apiladas o superpuestas.

La elección del tipo de gráfico adecuado dependerá del tipo de datos que se tengan, el objetivo del análisis y la audiencia a la que se dirige la presentación de los resultados.
